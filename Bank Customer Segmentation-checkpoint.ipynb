## Step 1: Preparing the Data

#### Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

#### 1. Read in the data
Reading in the Bank_Churn.csv file into Python and viewing the top 5 rows.

```python
data = pd.read_csv(r"OneDrive/AG_Portfolio/Bank Customer Segmentation/datasets/Bank_Churn.csv")
```

```python
data.head()
```

#### 2. Creating a subset of the data
Creating a Pandas DataFrame containing all feilds except CustomerId, Surname, and Exited.
CustomerId and Surname are unique identifiers for the row and can't be used for clustering. Exited will be overlaid on top of clusters later.

```python
data_subset = data.drop(columns=['CustomerId', 'Surname', 'Exited'])
data_subset.head()
```

#### 3. Making text fields numeric
Making all text fields numeric. Making a copy to not lose original subset dataset.

```python
data_clean = data_subset.copy()
data_clean.head()
```

```python
data_clean.Geography.value_counts()
```

```python
# Geography has 3 unique values, will use dummy variables.
```

```python
data_clean.Gender.value_counts()
```

```python
# Gender has 2 unique values, can assign 1's and 0's to column.
```

```python
data_clean['Gender'] = (
    np.where(
        data_clean.Gender == "Female",
        1,
        0
    )
)
```

```python
data_clean.head()
```

```python
data_clean = pd.get_dummies(data_clean, columns=['Geography'], dtype='int', prefix='', prefix_sep='')
```

```python
data_clean.head()
```

#### 4. Exploring the data
Examining statistics and distribution of each column.

```python
data_clean.describe()
```

```python
# Mean and range values for each column make sense.
```

```python
sns.pairplot(data_clean);
```

```python
# CreditScore: Looks approximately normally distributed over min/max possible credit score.
# Gender: Approximately evenly split. 45% female, 55% male.
# Age: Approximately normally distributed. Mean 39yo, StDev 10y.
# Tenure: Approximately uniformly distributed along tenure range.
# Balance: Large balance outlier ~250,000 balance. Besides that datapoint, approximately normally distributed.
# NumOfProducts: Essentially evenly split between 1 - 2 products. 3 - 4 number of products seem to be rare cases.
# HasCrCard: ~70% have credit card, ~30% don't. Not concerned since data set is relatively large (10K rows).
# IsActiveMember: Approximately evenly split. 51.5% active, 48.5% not active.
# EstimatedSalary: Uniformly distributed across salary range rather than normal, there is an even amount of representation for each salary class.
# France: ~50% of members from France. Majority of members from France. Remaining proportion of customers evenly split between Germany and Spain.
# Germany: ~25% of members from Germany.
# Spain: ~25% of members from Spain.
```

#### 5. Engineering new features
Engineering a new feature called ProductsPerYear

```python
data_clean.head()
```

```python
data_clean['ProductsPerYear'] = np.where(
    data_clean.Tenure == 0,
    data_clean.NumOfProducts,
    data_clean.NumOfProducts / data_clean.Tenure
)
```

```python
data_clean.describe()
```

## Step 2: Cluster the Data (Round 1)

#### 1. Scale the data using standardization
Standardize the data so that each column has a mean of 0 and a standard deviation of 1.

```python
from sklearn.preprocessing import StandardScaler
```

```python
data_clean.head()
```

```python
scaler = StandardScaler()
data_scaled = pd.DataFrame(scaler.fit_transform(data_clean), columns=data_clean.columns)
data_scaled.head()
```

```python
data_scaled.describe()
```

```python
# confirming all data is standardized with a mean of 0 and standard deviation of 1.
```

#### 2. Fitting K-Means model with 2-15 clusters
Fitting K-Means Clustering models on the standardized data with 2-15 clusters to create an inertia plot.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# writing a loop to fit models with 2 to 15 clusters
# creating an empty list to hold many inertia and silhouette values
inertia_values = []
silhouette_scores = []

# creating 2 - 15 clusters, and adding the inertia scores and silhouette scores to the lists
for k in range(2,16):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42) # changed n_init from auto to 10
    kmeans.fit(data_scaled)
    inertia_values.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(data_scaled, kmeans.labels_, metric='euclidean', sample_size=None))
```

#### 3. Plotting the inertia values and finding the elbow
Identifying the elbow of the inertia plot and fitting the K-Means model using the value of k.

```python
import matplotlib.pyplot as plt

# inertia series
inertia_series = pd.Series(inertia_values, index=range(2, 16))

# plot the data
inertia_series.plot(marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Number of Clusters vs. Inertia")
plt.show()
```

```python
# The elbow of the inertia plot occurs at 5 clusters. Will refit model with 5 clusters.
```

```python
kmeans5 = KMeans(n_clusters=5, n_init=10, random_state=42)
kmeans5.fit(data_scaled)
```

#### 3. Checking the number of customers in each cluster
Checking the number of customers in each cluster.

```python
from collections import Counter

Counter(kmeans5.labels_)
```

```python
# None of the clusters look extremely big or small so looks good to proceed.
```

#### 5. Creating a heat map of the cluster centers and interpreting the clusters
Creating a heat map of cluster centers and interpreting clusters.

```python
# creat a heat map of the cluster centers
clusters_centers5 = pd.DataFrame(kmeans5.cluster_centers_, columns=data_scaled.columns)

plt.figure(figsize=(10,2))
sns.heatmap(clusters_centers5, annot=True, cmap="RdBu", fmt=".1f", linewidths=.5);
```

* Cluster 0: These are the customers that sign up for many products in a short time. Noted by a relatively high amount of ProductsPerYear, NumOfProducts, and small Tenure.
* Cluster 1: These are generally French customers with few products and high balance. Noted by relatively high proportion of France, low NumOfProducts, and relatively high Balance.
* Cluster 2: These are generally German customers with a high balance.
* Cluster 3: These are generally French customers with more products and a low balance.
* Cluster 4: These are our Spain customers.

1 Cluster heavily dependent on ProductsPerYear. Remaining seem to be primarily Geography dependent (2 French customer clusters - 1 with few products/high balance and one with more products/low balance. 1 German customer with high balance, 1 spanish customer cluster.)

Next, will try to update the data model to make improved clusters.
